```python
import pandas as pd
import numpy as np
```

```python
DATA = pd.read_excel("input/sheet.xlsx", sheet_name='schedule')
# display(DATA.head(5))
DATA = DATA.drop([k for k in DATA.columns if k.startswith("Unnamed")], axis=1)
DATA['Day'] = DATA["Day"].fillna(method='ffill')
DATA.head(10)
DATA["name"] = DATA.apply(lambda x: (str(x["FirstName"]).strip() + str(x["LastName"]).strip()).lower(), axis=1)
DATA.head(10)
```

```python
TITLES = pd.read_excel("input/sheet.xlsx", sheet_name='titles')
display(TITLES.head(5))
TITLES = TITLES.dropna(subset="Surname", axis=0)
TITLES = TITLES.drop_duplicates(subset=["First name", "Surname"], keep='last')
TITLES = TITLES.drop([k for k in TITLES.columns if k.startswith("Unnamed")], axis=1)
TITLES["name"] = TITLES.apply(lambda x: (str(x["First name"]).strip() + str(x["Surname"]).strip()).lower(), axis=1)
TITLES.head(10)
```

```python
APPLICATIONS = pd.read_excel("input/sheet.xlsx", sheet_name='Applications')
display(APPLICATIONS.head(5))
APPLICATIONS = APPLICATIONS.dropna(subset="Surname", axis=0)
APPLICATIONS = APPLICATIONS[["Name", "Surname", "Abstract Title", "Abstract Body"]]
APPLICATIONS = APPLICATIONS.drop_duplicates(subset=["Name", "Surname"], keep='last')
APPLICATIONS = APPLICATIONS.drop([k for k in APPLICATIONS.columns if k.startswith("Unnamed")], axis=1)
APPLICATIONS["name"] = APPLICATIONS.apply(lambda x: (str(x["Name"]).strip() + str(x["Surname"]).strip()).lower(), axis=1)
APPLICATIONS.head(10)
```

```python
FINAL = pd.read_excel("input/sheet.xlsx", sheet_name='FINAL')
display(FINAL.head(5))
FINAL["name"] = FINAL.apply(lambda x: (str(x["Name, Surname, Special Name, Love Name"]) + str(x["Unnamed: 1"])).lower(), axis=1)
FINAL.head(10)
```

```python
def find_name(name):
    title = None
    abstract = None
    afil = None
    
    # first look for the title and abstract
    df = TITLES.loc[TITLES["name"] == name,:]
    if df.shape[0] > 0:
        title, abstract = df[["Talk title", "Short abstract"]].iloc[-1,:]
    if title is None:
        df = APPLICATIONS.loc[APPLICATIONS["name"] == name,:]
        if df.shape[0] > 0:
            title, abstract = df[["Abstract Title", "Abstract Body"]].iloc[-1,:]
        
    # find affiliation
    df = FINAL.loc[FINAL["name"] == name,:]
    if df.shape[0] > 0:
        afil = df["Affiliation"].iloc[-1]
        
    return (afil, title, abstract)
```

```python
from collections import defaultdict

for day, df in DATA.groupby("Day"):
    output = []
    
    df["Start"] = pd.to_datetime(df["Start"].astype(str).apply(pd.Timestamp))
    df["End"] = pd.to_datetime(df["End"].astype(str).apply(pd.Timestamp))
    df = df.sort_values(by="Start")
    df = df.dropna(subset="Start", axis=0).reset_index(drop=True)
#     df = df.dropna(subset="LastName", axis=0)
    
    #display(df.head(2))
    print(df.TalkType.unique())
    for _, r in df.iterrows():
#         display(r)
        dur = r["End"] - r["Start"]
        if r.TalkType.lower() in ("lunch", "break"):
            continue
        if r.TalkType.lower() in ("short", "regular", "spotlight"):
            is_talk = True
        else:
            is_talk = False
            
        key = r["name"]
        if is_talk:
            afil, title, abstract = find_name(key)
            if title != r["Title"]:
                if title is None:
                    print("No title found for:", key)
                    title = "TBA"
                else:
                    print(f"Warning, different titles for {key}: {title} \n vs \n {r.Title}")
            if afil is None:
                print("No affiliation found for:", key)
                afil = "TBA"
            
        d = {
            "Start" : r["Start"].strftime('%H:%M'),
            "End" : r["End"].strftime('%H:%M'),
            "Duration" : int(dur.components.hours*60+dur.components.minutes),
            "TalkDuration" : r["DurationStr"] if is_talk else "",
            "FirstName" : r["FirstName"],
            "LastName" : r["LastName"] if is_talk else "",
            "Title" : r["Title"] if is_talk else "",
            "TalkType": r["TalkType"].lower(),
            "Abstract" : abstract,
            "Affiliation" : afil
        }        
#         print(d)
        output.append(d)
    output = pd.DataFrame(output)
    #display(output)
    output.to_csv(f"csvs/{day.lower()}.csv", sep="?", header=False)
```

```python
from collections import defaultdict

for day, df in DATA.groupby("Day"):
    output = []
    
    df["Start"] = pd.to_datetime(df["Start"].astype(str).apply(pd.Timestamp))
    df["End"] = pd.to_datetime(df["End"].astype(str).apply(pd.Timestamp))
    df = df.sort_values(by="Start")
    df = df.dropna(subset="Start", axis=0).reset_index(drop=True)
#     df = df.dropna(subset="LastName", axis=0)
    
    for _, r in df.iterrows():
#         display(r)
        dur = r["End"] - r["Start"]
        is_break = False
        is_talk = False

        if r.TalkType.lower() in ("lunch", "break"):
            is_break = True
        elif r.TalkType.lower() in ("short", "regular", "spotlight"):
            is_talk = True
            
        key = r["name"]
        if is_talk:
            afil, title, abstract = find_name(key)
            if title != r["Title"]:
                if title is None:
                    print("No title found for:", key)
                    title = "TBA"
                else:
                    print(f"Warning, different titles for {key}: {title} \n vs \n {r.Title}")
            if afil is None:
                print("No affiliation found for:", key)
                afil = "TBA"
        elif is_break:
            typ = r["TalkType"]
            if typ == "lunch":
                r["Title"] = "Lunch"
            elif typ == "break":
                r["Title"] = "Break"
            
            r["TalkType"] = "break"
                

            
        if isinstance(r["LastName"], float):
            talk_uuid = "--"
        else:
            #print(r["LastName"])
            talk_uuid = unidecode.unidecode(r["LastName"]).lower()
        d = {
            "Start" : r["Start"].strftime('%H:%M'),
            "End" : r["End"].strftime('%H:%M'),
            "Duration" : int(dur.components.hours*60+dur.components.minutes),
            "TalkDuration" : r["DurationStr"] if is_talk else "",
            "FirstName" : r["FirstName"],
            "LastName" : r["LastName"] if is_talk else "",
            "Title" : r["Title"] if is_talk else "",
            "TalkType": r["TalkType"].lower(),
            "talkcode": talk_uuid,
            "Affiliation" : afil
        }        
        output.append(d)

        import unidecode

        from textwrap import dedent
        outdata = dedent(
           f"""
            <div class="cd-schedule-modal__event-info">
                <div> {abstract}
                </div>
            </div>
            """)
        
        with open(f"abstracts/{talk_uuid}.html", "w") as f:
            f.write(outdata)

    output = pd.DataFrame(output)
    #display(output)
    output.to_csv(f"csvs/{day.lower()}.csv", sep="?", header=False)
```
